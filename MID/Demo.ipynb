```python
import tensorflow as tf
import numpy as np

def predict_image_with_lenet(image_path):

  # Load the image
  img = tf.keras.preprocessing.image.load_img(image_path, color_mode='grayscale', target_size=(28, 28))
  # Convert to numpy array
  img_array = tf.keras.preprocessing.image.img_to_array(img)
  # Reshape for the model (add batch dimension and channel dimension)
  img_array = np.expand_dims(img_array, axis=0)
  img_array = np.expand_dims(img_array, axis=-1)
  # Normalize the image
  img_array = img_array / 255.0

  # Load the saved model
  model = tf.keras.models.load_model('/content/lenet_mnist_model.keras')

  # Make a prediction
  predictions = model.predict(img_array)

  return predictions
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf

def plot_image(image_path):
  """
  Loads an image and plots it using matplotlib.

  Args:
    image_path (str): The path to the image file.
  """
  # Load the image
  img = tf.keras.preprocessing.image.load_img(image_path, color_mode='grayscale')

  # Plot the image
  plt.imshow(img, cmap='gray')
  plt.axis('off')  # Turn off axis numbers and ticks
  plt.show()
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import os

def plot_image_with_prediction(image_path, prediction):
  # Load the image
  img = tf.keras.preprocessing.image.load_img(image_path, color_mode='grayscale')

  # Plot the image
  plt.imshow(img, cmap='gray')
  plt.axis('off')  # Turn off axis numbers and ticks
  plt.title(f"Predicted: {prediction}") # Add the predicted digit as the title
  plt.show()

# Get a list of all image files in the TestNumbers folder
image_files = [f for f in os.listdir('/content/TestNumbers') if f.endswith(('.png', '.jpg', '.jpeg'))]

for image_file in image_files:
  # Construct the full path
  full_image_path = "/content/TestNumbers/" + image_file
  predictions = predict_image_with_lenet(full_image_path)
  print(f"Predictions for {full_image_path}: {predictions}")

  # Get the predicted digit (index of the highest probability)
  predicted_digit = np.argmax(predictions)

  # Plot the image and display the predicted digit
  plot_image_with_prediction(full_image_path, predicted_digit)
```
